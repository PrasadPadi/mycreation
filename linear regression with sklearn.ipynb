```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wt=pd.read_csv('weight-height.csv')
wt.head()
```

```python
sns.distplot(wt['Height'],kde=True)
plt.title('distribution of Height ')
plt.show()
```

```python
sns.distplot(wt['Weight'],kde=True)
plt.title('distribution of weight ')
plt.show()
```

```python
sns.relplot('Weight','Height',kind='line',data=wt)
plt.scatter(wt['Weight'],wt['Height'])
plt.title('plot for relation b/n weight and height')
plt.show()
```

```python
sns.lmplot('Weight','Height',data=wt)
plt.title('LMplot for relation b/n weight and height')
plt.show()
```
