```python
def gradient_descent(x,y):
    m=0
    b=0
    cost_fn=0
    alpa=0.01
    n=len(x)
    mi=[]
    bi=[]
    cfn=[]
    for i in range(1000):
        dm=(-2/n)*np.sum(x*(y-(m*x+b)))
        db=(-2/n)*np.sum(y-(m*x+b))
        cost_fn=(1/n)*np.sum((y-(m*x+b))**2)
        mi.append(m)
        bi.append(b)
        cfn.append(cost_fn)
        m=m-alpa*dm
        b=b-alpa*db
        print(m,b,cost_fn)
    return mi,bi,cfn    
```
