```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x=np.arange(1,11)
y=np.arange(11,21)
print(x,y)
len(x)
```

```python
plt.plot(x,y)
plt.scatter(x,y)
plt.xlabel('values of x')
plt.ylabel('values of y')
plt.title('plot for x and y')
plt.show()
```

```python
def line_coef(x, y): 
    n = len(x) 
    mx=np.mean(x)
    my=np.mean(y) 
    sxy = np.sum(y*x) - n*my*mx 
    Sxx = np.sum(x*x) - n*mx*mx 
    m = sxy/Sxx 
    b = my -m*mx 
  
    return m,b 
```

```python
m,b=line_coef(x, y)
print('m:{}\nb:{}'.format(m,b))
```

```python
y_pred = b + m*x 
plt.scatter(x, y, color = "black")
plt.plot(x, y_pred, color = "red") 
plt.xlabel('values of x')
plt.ylabel('values of y_predicted')
plt.title('plot for x and y-predicted')
plt.show()
```
