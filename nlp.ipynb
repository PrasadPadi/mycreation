```python
dat['sentiment']=dat['sentiment'].replace({'Neutral':0,'Positive':1,'Negative':-1})
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
dat['text'].shape
```

```python
corpus=[]
for i in range(0,13871):
    text=re.sub('[^a-zA-Z]',' ',dat['text'][i])
    text=text.lower()
    text=text.split()
    text=[word for word in text if not word in stopword]
    text=[wn.lemmatize(word) for word in text]
    text=' '.join(text)
    corpus.append(text)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vec=TfidfVectorizer()
x=tfidf_vec.fit_transform(corpus)

tfidf_vec.get_feature_names()
```

```python
x.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec=CountVectorizer(max_features=5000)
x=count_vec.fit_transform(corpus)

count_vec.get_feature_names()
```

```python
x.shape
```

```python
x=x.toarray()
```

```python
y=dat['sentiment']
```

```python
y.head(n=10)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.naive_bayes import GaussianNB
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.20, random_state=0)
gnb = GaussianNB()
gnb.fit(x_train,y_train)
```

```python
y_pred=gnb.predict(x_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
ac=accuracy_score(y_test,y_pred)
ac
```

```python
from sklearn.linear_model import LogisticRegression
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.20, random_state=0)
gnb = LogisticRegression()
gnb.fit(x_train,y_train)
y_pred=gnb.predict(x_test)
```

```python
ac=accuracy_score(y_test,y_pred)
ac
```

```python
from sklearn.tree import DecisionTreeClassifier
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.20, random_state=0)
gnb =  DecisionTreeClassifier()
gnb.fit(x_train,y_train)
y_pred=gnb.predict(x_test)
```

```python
ac=accuracy_score(y_test,y_pred)
ac
```

```python
from sklearn.ensemble import RandomForestClassifier
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.20, random_state=0)
gnb = RandomForestClassifier()
gnb.fit(x_train,y_train)
y_pred=gnb.predict(x_test)
```

```python
ac=accuracy_score(y_test,y_pred)
ac
```
