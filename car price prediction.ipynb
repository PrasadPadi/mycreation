```python
def estimated_coff(x,y):
    n=np.size(x)
    ymean=np.mean(x)
    xmean=np.mean(y)
    ssxy=np.sum(y*x) - n*ymean*xmean
    ssxx=np.sum(x*x) - n*xmean*xmean
    m=ssxy/ssxx
    c=ymean-m*xmean
    return m,c
```

```python
x=car['engine-size']
y=car['price']
b=estimated_coff(x,y)
print("Estimated coefficients:\n m= {} \n c = {}".format(b[0], b[1])) 
```

```python
plt.scatter(x, y, color = "b")
y1=b[1]*x+b[0]
plt.plot(x,y1,color = "k")
plt.xlabel('Engine size')
plt.ylabel('Price')
plt.show()
```
